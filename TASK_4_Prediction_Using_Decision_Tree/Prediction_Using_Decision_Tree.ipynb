# DECISION TREE

```python
#Setting The x and Y Labels 
x = dset[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
y = dset['Species'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state=1)
```

```python
# Build Decision tree Model 
from sklearn.tree import DecisionTreeClassifier
destree = DecisionTreeClassifier(criterion='entropy',random_state=1)
destree.fit(x_train,y_train)
```

```python
print(destree.score(x_test,y_test))
```

# Hence we can conclude that the Accuracy is 95 percent

# Decision Tree

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
fig = plt.figure(figsize=(200,200))
_ = tree.plot_tree(destree, 
                   feature_names=x,  
                   class_names=y,
                   filled=True)
```
