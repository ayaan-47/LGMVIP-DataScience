# DATA VISUALIZATION
# To get the maximum/minimum and other values for Each particular category


```python
print(dset['ATTACK_TYPE'].value_counts().idxmax()," ----> is the most used Attack Type")
print(dset['country_txt'].value_counts().idxmax()," ----> was attacked the most number of times")
print(dset['STATE'].value_counts().index[1]," ----> is the most attacked city") #as first entry is 'unknown'
print("Region which is attacked the most is :",dset['REGION'].value_counts().idxmax())
print("Year which was attacked the most is  :",dset['YEAR'].value_counts().idxmax())

m = dset['MONTH'].value_counts().idxmax()
month = "none"
if m == 1 :
    month = "January"
elif m == 2 : 
    month = "February"
elif m == 3 : 
    month = "March"
elif m == 4 : 
    month = "April"
elif m == 5 : 
    month = "May"
elif m == 6 : 
    month = "June"
elif m == 7 : 
    month = "July"
elif m == 8 : 
    month = "August"
elif m == 9 : 
    month = "September"
elif m == 10 : 
    month = "October"
elif m == 11 : 
    month = "November"
elif m == 12 : 
    month = "December"
print(f"Highest Attacks were noted in the month of  {month}")
print("Maximum number of attacks were placed by ",dset['GROUP'].value_counts().index[1], "Group")
```

```python
dset['YEAR'].value_counts(dropna = False).sort_index()
```

# Terrorist Activities Each Year 

```python

x_year = dset['YEAR'].unique()
y_count_years = dset['YEAR'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'Blues')
plt.xticks(rotation = 50)
plt.xlabel('Year of Attack ')
plt.ylabel('Attacks per Year')
plt.title('How many years till the attack')
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot('YEAR',data=dset,palette='Reds',edgecolor=sns.color_palette("YlOrBr", 10))
plt.xticks(rotation=50)
plt.title('Total Terrorist attacks or activities each year')
plt.show()
```

# Terrorist Activities By Region in Each Year

```python


pd.crosstab(dset.YEAR, dset.REGION).plot(kind='area',figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
dset['WOUNDED'] = dset['WOUNDED'].fillna(0).astype(int)
dset['KILLED'] = dset['KILLED'].fillna(0).astype(int)
dset['Casualities'] = dset['KILLED'] + dset['WOUNDED']
```

# 10 Worst Terrorist Attacks

```python
# 10 Worst Terrorist Attacks
dset1 = dset.sort_values(by='Casualities',ascending=False)[:10]
```

```python

heat=dset1.pivot_table(index='COUNTRY',columns='YEAR',values='Casualities')
heat.fillna(0,inplace=True)
```

# How Many People Were Killed in the Attacks

```python
kd = dset.loc[:,'KILLED']
print(int(sum(kd.dropna())))
```

# Types of attacks and total deaths caused by Them 

```python
ad = dset.loc[:,'ATTACK_TYPE']
kdt = pd.concat([ad,kd],axis=1)
```

```python
kdtf = kdt.pivot_table(columns = 'ATTACK_TYPE',values='KILLED',aggfunc='sum')
kdtf
```

```python
# Total number of People Killed In the Attacks(Country wise list)
cd = dset.loc[:,'country_txt']
kcd = pd.concat([cd,kd],axis = 1)
kcdf = kcd.pivot_table(columns='country_txt',values='KILLED',aggfunc = 'sum')
kcdf
```

# GRAPH for total number of People killed in the Attacks(Country Wise List)

```python
#GRAPH for total number of People killed in the Attacks(Country Wise List)
labels = kcdf.columns.tolist()
labels = labels[152:206]
index = np.arange(len(labels))
transpoze = kcdf.T
values = transpoze.values.tolist()
values = values[152:206]
values = [int(i[0]) for i in values]
colors = ['green','black','grey','red','blue','pink','yellow','violet','cyan','orange']
fig, ax = plt.subplots(1, 1)
ax.yaxis.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
plt.bar(index, values, color = colors, width = 0.9)
plt.ylabel('TOTAL_KILLED', fontsize=20)
plt.xlabel('COUNTRY_NAME', fontsize = 20)
plt.xticks(index, labels, fontsize=20, rotation=90)
plt.title('Total Number of People Killed', fontsize = 20)
plt.show()
```
