```python
import cv2
import matplotlib.pyplot as plt
pic = cv2.imread("Doggy_Panther.jpg")
picture = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB) # I have used cv2.cvtColor to read/show the image in rgb colors

plt.imshow(picture)
```

# Converting Image to Grey and then Inverting it

```python
grey_panther = cv2.cvtColor(picture,cv2.COLOR_RGB2GRAY) # To Convert it into Grey 

inverted_panther = cv2.bitwise_not(grey_panther)#The bitwise_not operator inverts/flips or reverses all the pixels of the image
plt.imshow(inverted_panther)

```

# Making the image blur and then again inverting it. 

```python
blurr_panther = cv2.GaussianBlur(inverted_panther,(21,21),0)


inverted_bp = cv2.bitwise_not(blurr_panther)
plt.imshow(inverted_bp)

```

# Finally Converting the inverted blurr image into the Pencil Sketch Image

```python
pencil_panther = cv2.divide(grey_panther,inverted_bp,scale = 256.0)


plt.imshow(pencil_panther,cmap="gray") 
```
